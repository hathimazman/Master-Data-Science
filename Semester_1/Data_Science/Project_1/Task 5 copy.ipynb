# Unary Functions

```python
def unary_fx(x : np.array):
    '''
    a function that runs all unary functions on a given array
    
    abs() and fabs() : Compute the absolute value element-wise for integer, floatingpoint,
                        or complex values
    sqrt()           : Compute the square root of each element
    square()         : Compute the square of each element
    exp()            : Compute the exponent of each element
    log()            : Compute the natural logarithm of each element
    log10()          : Compute the common logarithm of each element
    log2()           : Compute the binary logarithm of each element
    log1p()          : Compute the natural logarithm of one plus each element
    sign()           : Compute the sign of each element: 1 (positive), 0 (zero), or -1 (negative)
    ceil()           : Compute the ceiling of each element
    floor()          : Compute the floor of each element
    rint()           : Round elements to the nearest integer
    modf()           : Return fractional and integral parts of array as separate array
    isnan()          : Return True if the value is NaN, False otherwise
    isfinite()       : Return True if the value is finite, False otherwise
    isinf()          : Return True if the value is infinity, False otherwise
    cos()            : Compute the cosine of each element
    cosh()           : Compute the hyperbolic cosine of each element
    sin()            : Compute the sine of each element
    sinh()           : Compute the hyperbolic sine of each element
    tan()            : Compute the tangent of each element
    tanh()           : Compute the hyperbolic tangent of each element
    arccos()         : Compute the inverse cosine of each element
    arccosh()        : Compute the inverse hyperbolic cosine of each element
    arcsin()         : Compute the inverse sine of each element
    arcsinh()        : Compute the inverse hyperbolic sine of each element
    arctan()         : Compute the inverse tangent of each element
    arctanh()        : Compute the inverse hyperbolic tangent of each element
    logical_not()    : Compute the truth value of not x element-wise
    '''
    unary_fx_dict ={
        'Absolute Value': np.abs(x),
        'Float Abs Value': np.fabs(x),
        'Square Root': np.sqrt(x),
        'Square': np.square(x),
        'Exponential': np.exp(x),
        'Natural log': np.log(x),
        'log10': np.log10(x),
        'log2': np.log2(x),
        'log(1+x)': np.log1p(x),
        'Sign': np.sign(x),
        'Ceiling': np.ceil(x),
        'Floor': np.floor(x),
        'Round nearest Int': np.rint(x),
        'Fractional and Integral': np.modf(x),
        'is NaN': np.isnan(x),
        'is finite': np.isfinite(x),
        'is inf': np.isinf(x),
        'cos': np.cos(x),
        'cosh': np.cosh(x),
        'sin': np.sin(x),
        'sinh': np.sinh(x),
        'tan': np.tan(x),
        'tanh': np.tanh(x),
        'arccos': np.arccos(x),
        'arccosh': np.arccosh(x),
        'arcsin': np.arcsin(x),
        'arcsinh': np.arcsinh(x),
        'arctan': np.arctan(x),
        'arctanh': np.arctanh(x),
        'logical_not': np.logical_not(x)
    }
    return unary_fx_dict
```

```python
def random_choice(x : list, y : int, fx):
    '''
    a function that takes in an array and a number and returns number of random choice of unary function

    Parameters:
    x : list : list of  array
    y : int : number of random choices to return
    fx : function : a function that runs all functions on a given array
    '''
    if len(x) > 1:
        for b in x:
            unary_fx_dict = fx(b)
            for i in range(y):
                random_choice = np.random.choice(list(unary_fx_dict.keys()), replace=False)
                print(f"""
{random_choice}:
{unary_fx_dict[random_choice]}
                    """)
            print(f'{'*'*50}')
    else:
        unary_fx_dict = fx(x)
        for i in range(y):
            random_choice = np.random.choice(list(unary_fx_dict.keys()), replace=False)
            print(f"""
{random_choice}:
{unary_fx_dict[random_choice]}
                    """)
            print(f'{'*'*50}')
        
```

# Binary Universal Functions

```python
def BUN_fx(x : list):
    '''
    a function that runs all binary universal functions on a given array

    add()                       : Add corresponding elements in arrays
    subtract()                  : Subtract elements in second array from first array
    multiply()                  : Multiply array elements
    divide() and floor_divide() : Divide first array by second array
    power()                     : Raise elements in first array to powers indicated in second array
    maximum(), fmax()           : Element-wise maximum between two arrays, ignoring any NaNs
    minimum(), fmin()           : Element-wise minimum between two arrays, ignoring any NaNs
    mod()                       : Element-wise modulus (remainder of division)
    copysign()                  : Copy sign of values in second argument to values in first argument
   
    greater(), greater_equal(), 
    less(), less_equal(), 
    equal(), not_equal()        : Perform element-wise comparison
    
    logical_and(), logical_or(),
    logical_xor()               : Perform element-wise logical operations
    '''
    try:
        binary_fx_dict = {
            'Addition': np.add(x[0], x[1]),
            'Subtraction': np.subtract(x[0], x[1]),
            'Multiplication': np.multiply(x[0], x[1]),
            'Division': np.divide(x[0], x[1]),
            'Floor Division': np.floor_divide(x[0], x[1]),
            'Power': np.power(x[0], x[1]),
            'Maximum': np.maximum(x[0], x[1]),
            'Minimum': np.minimum(x[0], x[1]),
            'Modulus': np.mod(x[0], x[1]),
            'Copy Sign': np.copysign(x[0], x[1]),
            'Greater': np.greater(x[0], x[1]),
            'Greater Equal': np.greater_equal(x[0], x[1]),
            'Less': np.less(x[0], x[1]),
            'Less Equal': np.less_equal(x[0], x[1]),
            'Equal': np.equal(x[0], x[1]),
            'Not Equal': np.not_equal(x[0], x[1]),
            'Logical And': np.logical_and(x[0], x[1]),
            'Logical Or': np.logical_or(x[0], x[1]),
            'Logical Xor': np.logical_xor(x[0], x[1])
        }
        return binary_fx_dict
    except ValueError as e:
        print(e)
    
```

# Basic array statistical methods

```python
def BAS_methods(x : np.array):
    temp = pd.DataFrame()
    '''
    a function that runs all basic array statistical methods on a given array
    
    sum()                   : Sum of all the elements in the array or along an axis; zero-length arrays have sum 0
    mean()                  : Arithmetic mean; zero-length arrays have NaN mean
    std() and var()         : Standard deviation and variance, respectively, with optional degrees of freedom adjustment
                              (default denominator n)
    min()and max()          : Minimum and maximum
    argmin() and argmax()   : Indices of minimum and maximum elements, respectively
    cumsum()                : Cumulative sum of elements starting from 0
    cumprod()               : Cumulative product of elements starting from 1
    '''
    temp['Shape'] = [x.shape]
    temp['Sum'] = [x.sum()]
    temp['Mean'] = [x.mean()]
    temp['Standard Deviation'] = [x.std()]
    temp['Variance'] = [x.var()]
    temp['Minimum'] = [x.min()]
    temp['Maximum'] = [x.max()]
    temp['ArgMin'] = [x.argmin()]
    temp['ArgMax'] = [x.argmax()]
    temp['CumSum'] = [x.cumsum()]
    temp['CumProd'] = [x.cumprod()]
    return temp
```

# Commonly used `numpy.linalg` functions

```python
def linalg_func(x,**kwargs):
    '''
    a function that runs all linear algebra functions on a given array

    diag()            : Return the diagonal (or off-diagonal) elements of a square matrix as a 1D array, or convert a 1D array into a square matrix with zeros on the off-diagonal
    dot()             : Matrix multiplication
    trace()           : Compute the sum of the diagonal elements
    det()             : Compute the matrix determinant
    eig()             : Compute the eigenvalues and eigenvectors of a square matrix
    inv()             : Compute the inverse of a square matrix
    pinv()            : Compute the Moore-Penrose pseudo-inverse of a matrix
    qr()              : Compute the QR decomposition
    svd()             : Compute the singular value decomposition (SVD)
    solve()           : Solve the linear system Ax = b for x, where A is a square matrix
    lstsq()           : Compute the least-squares solution to Ax = b
    '''
    for key, value in kwargs.items():
        if x.shape[1] == value.shape[1]:
            linal_fx_dict = {
                'Diagonal': np.diag(x),
                'Dot': np.dot(x, value),
                'Trace': np.trace(x),
                'Determinant': np.linalg.det(x),
                'Eigenvalues': np.linalg.eig(x),
                'Inverse': np.linalg.inv(x),
                'Pseudo Inverse': np.linalg.pinv(x),
                'QR': np.linalg.qr(x),
                'SVD': np.linalg.svd(x),
                'Solve': np.linalg.solve(x, value),
                'Least Squares': np.linalg.lstsq(x, value)
            }
        return linal_fx_dict
    else:
        print("Arrays are not of the same shape")
```

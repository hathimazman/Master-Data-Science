```python
#3)
def count_char(file_paths):
    for file_path in file_paths:
        print(f'For {file_path}:')
        all_letters = {}
        with open(file_path, 'r') as file:
            content = file.read().lower()
            print('This is the number of characters: ' + str(len(content)))
            # returns the number of characters in the file
            for letter in content:
                if letter in all_letters:
                    all_letters[letter] += 1 
                else:
                    all_letters[letter] = 1
            for letter, freq in all_letters.items():
                print(f"{letter}: {freq}")

count_char(["text1.txt","text2.txt","text3.txt"])
```

```python
def word_freq(file_paths):
    all_words = []
    word_frequencies={}
    # Loop through each file path
    for file_path in file_paths:
        with open(file_path, 'r',encoding='UTf=8') as file:
            content = file.read()
            words = (content.lower())  # Convert to lowercase
            for word in words:
                if word.isalpha() and word not in STOPWORDS:
                # Check if the word is already in the dictionary
                    if word in word_frequencies:
                        word_frequencies[word] += 1  # Increment the count
                    else:
                        word_frequencies[word] = 1
    
    # Return the word frequencies
    return word_frequencies

# Example usage:
file_paths = ["01Alice_Wonderland.txt", "02Little_Wizard_Oz.txt","03Casebook_Sherlock_Holmes.txt"]
word_frequencies = word_freq(file_paths)

# Print the word frequencies
for word, freq in word_frequencies.items():
    print(f"{word}: {freq}")
```

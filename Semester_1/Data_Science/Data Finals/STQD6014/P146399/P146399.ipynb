```python
letter_txt1={}
with open(txt1,"r",encoding="UTF-8")as t:
    for line in t:
        for char in line.lower():
            if char.isalpha():
                letter_txt1[char]=letter_txt1.get(char,0)+1

select_txt1=sorted(letter_txt1.items(),key=lambda x:x[1],reverse=True)
print("frequebcy:freq.")
for letter,freq in select_txt1:
    print(letter,freq)
```

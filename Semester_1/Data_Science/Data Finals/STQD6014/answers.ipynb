```python
file_path1 = "./Data/PartB_Q3a/text1.txt"
file_path2 = "./Data/PartB_Q3a/text2.txt"
file_path3 = "./Data/PartB_Q3a/text3.txt"
file_paths = [file_path1,file_path2,file_path3]
```

```python
def count_letters(path):
    for num,path in enumerate(file_paths):
        with open(path, 'r') as file:
            content = file.read().lower()
            print(f"\nThe total number of chr in this file text{num+1}.txt is {len(content)}")
            
            letters = {}
            for letter in content:
                letter = letter.strip()
                if letter in letters:
                    letters[letter] += 1
                else:
                    letters[letter] = 1
            
            for let, count in letters.items():
                print(f"{let} : {count}")


count_letters(file_paths)
```

```python
path1 = "./Data/PartB_Q3b/01Alice_Wonderland.txt"
path2 = "./Data/PartB_Q3b/02Little_Wizard_Oz.txt"
path3 = "./Data/PartB_Q3b/03Casebook_Sherlock_Holmes.txt"
paths = [path1,path2,path3]
```

```python
full_text = ""
for path in paths:
    with open(path, 'r', encoding='utf-8-sig') as text:
        full_text += text.read().lower() + " "

print(f"First 300 characters of the combined file is : \n{full_text[:300]}")
```

```python
def preprocess(file):
    file = file.lower()

    # Remove punctuation
    punctuations = '''!()-[]{};:'"\,<>./?@#$%^&*_~'''
    for char in file:
        if char in punctuations:
            file = file.replace(char, "")

    # Remove numbers
    numbers = '0123456789'
    for num in numbers:
        file = file.replace(num, "")

    # Remove whitespaces
    file = file.strip()

    # Split the text into words
    words = file.split()

    return words

words = preprocess(full_text)
filtered_words = words.copy()
stop_words = ['the', 'and', 'is', 'in', 'to', 'with', 'a', 'of', 'for', 'on', 'it', 'this', 'that', 'you', 'i', 'he', 'she', 'we', 'they', 'as', 'at', 'by', 'an', 'be', 'not']
for word in words:
    if word in stop_words:
        filtered_words.remove(word)

print(f"First 50 words of the combined file after preprocessing are : \n{filtered_words[:50]}")
print(f"Total number of words in the combined file after preprocessing are : {len(filtered_words)}")
```

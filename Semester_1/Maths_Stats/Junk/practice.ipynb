### **Question 3: Shopping Habits**  
a) **P(Male and No Opinion):** \( \frac{50}{600} = 0.0833 \).  
b) **P(Female | Yes):** \( \frac{110}{200} = 0.55 \).  
c) **P(Male or Yes):** \( P(Male) + P(Yes) - P(Male and Yes) = \frac{300}{600} + \frac{200}{600} - \frac{90}{600} = 0.6833 \).  
d) **Independence of "Male" and "No":** Not independent, as \( P(Male | No) \neq P(Male) \).  
e) **Mutual Exclusivity of "Yes" and "No Opinion":** Mutually exclusive, as \( P(Yes \cap No\ Opinion) = 0 \).

### **Question 4: Sock Drawer**

c) **P(Unmatched Socks):** \( 1 - P(Red Pair) - P(White Pair) = 0.5053 \).
